# End-to-End EDA Project: Tips Dataset
In this project, we will explore the 'Tips' dataset to understand tipping behavior in restaurants.


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Load Data


```python
df = sns.load_dataset('tips')
df.head()
```

## 2. Initial Exploration


```python
df.info()
df.describe()
```

## 3. Univariate Analysis


```python
# Distribution of Total Bill
sns.histplot(df['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
# Count of Smokers vs Non-Smokers
sns.countplot(x='smoker', data=df)
plt.title('Count of Smokers')
plt.show()
```

## 4. Bivariate Analysis


```python
# Relationship between Total Bill and Tip
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
# Tips by Gender
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Tips by Gender')
plt.show()
```

## 5. Multivariate Analysis


```python
# Tips by Day and Smoker status
sns.violinplot(x='day', y='tip', hue='smoker', split=True, data=df)
plt.title('Tips by Day and Smoker')
plt.show()
```

## 6. Correlation Matrix


```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

## 7. key Insights
- Positive correlation between total bill and tip.
- Distribution of bills is right-skewed.
- Smokers vs Non-smokers tipping behavior can be further analyzed.
