# Matplotlib Basics
Master the foundation of plotting in Python.


```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plots


```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.show()
```

## Scatter Plots


```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y)
plt.show()
```

## Bar Charts


```python
plt.bar(['A', 'B', 'C'], [1, 3, 2])
plt.show()
```

## Histograms


```python
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30)
plt.show()
```

## Customization


```python
plt.plot(x, np.sin(x), label='Sine')
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
