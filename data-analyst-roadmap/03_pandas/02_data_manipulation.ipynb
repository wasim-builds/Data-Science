# Data Manipulation with Pandas
Learn how to filter, sort, and clean your data.


```python
import pandas as pd
import numpy as np
```

```python
# Create a sample DataFrame
df = pd.DataFrame({
    'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
    'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
    'C': np.random.randn(8),
    'D': np.random.randn(8)
})
```

## Filtering


```python
# Boolean indexing
df[df['A'] == 'foo']
```

```python
# Multiple conditions
df[(df['A'] == 'foo') & (df['C'] > 0)]
```

## Missing Data


```python
# Introduce missing data
df.loc[0, 'C'] = np.nan
df.loc[1, 'D'] = np.nan
df
```

```python
# Check for missing values
df.isna()
```

```python
# Drop rows with missing values
df.dropna()
```

```python
# Fill missing values
df.fillna(value=0)
```

## Sorting


```python
df.sort_values(by='C')
```
