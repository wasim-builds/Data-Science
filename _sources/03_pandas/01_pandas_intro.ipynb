# Introduction to Pandas
Pandas is a powerhouse tool for data manipulation and analysis. It's built on top of NumPy and provides easy-to-use data structures and data analysis tools.

## Key Concepts
- **Series**: One-dimensional labeled array.
- **DataFrame**: Two-dimensional labeled data structure (like a table).


```python
import pandas as pd
import numpy as np
```

## Creating Data Structures


```python
# Creating a Series
s = pd.Series([1, 3, 5, np.nan, 6, 8])
print(s)
```

```python
# Creating a DataFrame
dates = pd.date_range('20230101', periods=6)
df = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=list('ABCD'))
df
```

## Viewing Data


```python
df.head()  # View top rows
```

```python
df.tail(3) # View bottom rows
```

```python
df.index
```

```python
df.columns
```

```python
df.describe() # Rapid statistic summary
```

## Selection


```python
# Select a single column
df['A']
```

```python
# Slice rows
df[0:3]
```

```python
# Selection by Label (loc)
df.loc[dates[0]]
```

```python
# Selection by Position (iloc)
df.iloc[3]
```
