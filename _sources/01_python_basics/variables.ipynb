## 1. Variables

Variables are containers for storing data. In Python, you don't need to declare the type.

```python
# Creating variables
name = "Alice"
age = 25
height = 5.6
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
```

### Variable Naming Rules

âœ… **Good practices:**
- Use descriptive names: `user_age` instead of `a`
- Use snake_case for variables: `first_name`
- Start with letter or underscore

âŒ **Avoid:**
- Starting with numbers: `1name`
- Using spaces: `first name`
- Using reserved keywords: `class`, `for`, `if`

```python
# Good variable names
user_age = 30
total_price = 99.99
is_active = True

# Multiple assignment
x, y, z = 1, 2, 3
print(f"x={x}, y={y}, z={z}")

# Same value to multiple variables
a = b = c = 0
print(f"a={a}, b={b}, c={c}")
```

## 2. Data Types

Python has several built-in data types:

### 2.1 Integers (int)

Whole numbers, positive or negative

```python
# Integer examples
count = 100
temperature = -5
population = 1000000

print(type(count))  # Check the type
print(f"Count: {count}")
```

### 2.2 Floats (float)

Numbers with decimal points

```python
# Float examples
price = 19.99
pi = 3.14159
scientific = 2.5e3  # 2.5 * 10^3 = 2500

print(type(price))
print(f"Scientific notation: {scientific}")
```

### 2.3 Strings (str)

Text data enclosed in quotes

```python
# String examples
single_quote = 'Hello'
double_quote = "World"
multiline = """This is a
multiline string"""

# String concatenation
greeting = single_quote + " " + double_quote
print(greeting)

# String repetition
print("Ha" * 3)  # HaHaHa
```

```python
# String methods
text = "  Data Analysis  "

print(text.upper())      # Convert to uppercase
print(text.lower())      # Convert to lowercase
print(text.strip())      # Remove whitespace
print(text.replace("Data", "Python"))  # Replace text
print(len(text))         # Length of string
```

```python
# String formatting
name = "Bob"
age = 30

# Method 1: f-strings (recommended)
message1 = f"My name is {name} and I am {age} years old"

# Method 2: .format()
message2 = "My name is {} and I am {} years old".format(name, age)

# Method 3: % operator (old style)
message3 = "My name is %s and I am %d years old" % (name, age)

print(message1)
```

### 2.4 Booleans (bool)

True or False values

```python
# Boolean examples
is_raining = True
is_sunny = False

print(type(is_raining))

# Boolean from comparisons
print(5 > 3)      # True
print(10 == 5)    # False
print("a" < "b")  # True (alphabetical)
```

## 3. Type Conversion

Converting between data types

```python
# String to Integer
age_str = "25"
age_int = int(age_str)
print(f"Type: {type(age_int)}, Value: {age_int}")

# String to Float
price_str = "19.99"
price_float = float(price_str)
print(f"Type: {type(price_float)}, Value: {price_float}")

# Integer to String
count = 100
count_str = str(count)
print(f"Type: {type(count_str)}, Value: {count_str}")

# Float to Integer (truncates decimal)
pi = 3.14159
pi_int = int(pi)
print(f"Pi as int: {pi_int}")
```

## 4. Basic Operations

### 4.1 Arithmetic Operations

```python
# Basic arithmetic
a = 10
b = 3

print(f"Addition: {a + b}")           # 13
print(f"Subtraction: {a - b}")        # 7
print(f"Multiplication: {a * b}")     # 30
print(f"Division: {a / b}")           # 3.333...
print(f"Floor Division: {a // b}")    # 3
print(f"Modulus: {a % b}")            # 1
print(f"Exponentiation: {a ** b}")    # 1000
```

### 4.2 Comparison Operations

```python
x = 5
y = 10

print(f"Equal: {x == y}")              # False
print(f"Not equal: {x != y}")          # True
print(f"Greater than: {x > y}")        # False
print(f"Less than: {x < y}")           # True
print(f"Greater or equal: {x >= y}")   # False
print(f"Less or equal: {x <= y}")      # True
```

### 4.3 Logical Operations

```python
# AND, OR, NOT
is_adult = True
has_license = False

print(f"AND: {is_adult and has_license}")   # False
print(f"OR: {is_adult or has_license}")     # True
print(f"NOT: {not is_adult}")               # False
```
